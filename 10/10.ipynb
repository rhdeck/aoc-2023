# Part 1

```python
source = open('10.txt','r').read()
lines = source.split('\n')
rows = [list(line) for line in lines]
pipes = {
    "|": [(0,1), (0,-1)],
    "-": [(1,0), (-1,0)],
    "L": [(0,-1), (1,0)],
    "J": [(0,-1), (-1,0)],
    "7": [(0,1), (-1,0)],
    "F": [(0,1), (1,0)]
}
def locate_letter(letter, map):
    for y in range(len(map)):
        for x in range(len(map[y])):
            if map[y][x] == letter:
                return (x,y)
    return None

s_location = locate_letter("S", rows)
def handle_letter(coords, map, branchLetter = "S"):
    x,y = coords
    letter = map[y][x]
    if(letter == branchLetter):
        # Give me all possible exits
        exits = []
        for coordinates in [(x+1,y), (x-1,y), (x,y+1), (x,y-1)]:
            if(coordinates[0] < 0 or coordinates[1] < 0 or coordinates[0] >= len(map[0]) or coordinates[1] >= len(map)):
                continue
            if map[coordinates[1]][coordinates[0]] in pipes:
                #does this connect?
                outlets = pipes[map[coordinates[1]][coordinates[0]]]
                for outlet in outlets:
                    if (x-outlet[0], y-outlet[1]) == coordinates:
                        exits.append(coordinates)
        return exits
    elif letter in pipes:
        return [[x + outlet[0], y + outlet[1]] for outlet in pipes[letter]]
    return []

distances = {}

def save_distance(coords, length):
    coordstring = str(list(coords))
    distances[coordstring] = length

def check_distance(coords):
    coordstring = str(list(coords))
    return distances[coordstring] if coordstring in distances else None

def printMap(map):
    for row in map:
        print("".join(row))

outlets = handle_letter(s_location, rows)
distance = 0
save_distance(s_location, 0)
while len(outlets) > 0:
    
    next_outlets = []
    for outlet in outlets:
        if check_distance(outlet) == None:
            save_distance(outlet, distance)
            next_outlets += handle_letter(outlet, rows)
    if len(next_outlets) > 0:
        distance += 1 
    outlets = next_outlets
# print(distances)
"Part 1", distance
```

```python
def split_map(map, separator = ' '):
    out = []
    for row in map:
        joined = separator.join(row)
        out.append(list(joined))
        out.append(list(' ' * len(joined)))
    return out

def add_pipes(map, distances):
    for coordstring in distances:
        coords = eval(coordstring)
        # print(coords)
        x,y = coords
        x *= 2
        y *= 2
        # print(x,y)
        letter = map[y][x]
        # print(letter)
        # print(pipes)
        map[y][x] = "P"
        if letter in pipes:
            for outlet in pipes[letter]:
                map[y + outlet[1]][x + outlet[0]] = "P"   
    return map
empties = {}

def test_empty(coords, map):
    x,y = coords
    if  x >= 0 and y >= 0 and x < len(map[0]) and y < len(map) and map[y][x] not in ('P', 'O'):
        map[y][x] = "O"
        empties[str(coords)] = True
        return True
    return False   

def test_empty_neighbors(coords, map):
    out = []
    x,y = coords
    for outlet in [(1,0), (-1,0), (0,1), (0,-1)]:
        if test_empty((x + outlet[0], y + outlet[1]), map):
            out.append((x + outlet[0], y + outlet[1]))
    return out

def get_empty_perimeter(map):
    starts = []
    for y in range(len(map)):
        if y==0 or y == len(map) - 1:
            for x in range(len(map[y])):
                if test_empty((x,y), map):
                    starts.append((x,y))
                    # test_empty_neighbors((x,y), map)
        else:
            if test_empty((0,y), map):
                starts.append((0,y))
                # test_empty_neighbors((0,y), map)
            if test_empty((len(map[y]) - 1,y), map):
                starts.append((len(map[y]) - 1,y))
                # test_empty_neighbors((len(map[y]) - 1,y), map)
    return starts


def get_insides(map):
    line = ''.join([''.join(row) for row in map])
    l = len(line)
    fl = sum([1 if x not in (' ', 'O', 'P') else 0 for x in list(line)])
    return  fl

def print_map(map):
    for row in map:
        print("".join(row))


splitted = split_map(rows)
add_pipes(splitted, distances)
# print_map(splitted)
starts = get_empty_perimeter(splitted)
while len(starts) > 0:
    next_starts = []
    for start in starts:
        # print(start)
        next = test_empty_neighbors(start, splitted)
        next_starts += next
    # print(next_starts)
    starts = next_starts

print("Part 2", get_insides(splitted))
print_map(splitted)
```
